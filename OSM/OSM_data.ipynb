```python
import osmnx as ox #OSMnx: Python for street networks. Retrieve, model, analyze, and visualize street networks and other spatial data from OpenStreetMap.
import folium #Manipulate your data in Python, then visualize it in on a Leaflet map via folium.
from IPython.core.display import display, HTML

def folium_deepnote_show(m):
    data = m.get_root().render()
    data_fixed_height = data.replace('width: 100%;height: 100%', 'width: 100%').replace('height: 100.0%;', 'height: 609px;', 1)
    display(HTML(data_fixed_height))
```

Pełna lista kategorii dostępnych z OSM,
dostępna tu:https://wiki.openstreetmap.org/wiki/Map_features

```python
#Przykładowe użycie danych z osm

#wypożyczalnie rowerowe

place = 'Warsaw, PL'
tags = {'amenity': 'bicycle_rental'}

bicycle_rental = ox.geometries_from_place(place, tags=tags)
bicycle_rental = bicycle_rental[bicycle_rental.geom_type == 'Point'][:]
#Przykład danych
print(bicycle_rental[:3])
```

```python
#Mapa
m = folium.Map([52.2, 20.9], zoom_start=10)
locs = zip(bicycle_rental.geometry.y, bicycle_rental.geometry.x)
for location in locs:
    folium.CircleMarker(location=location,radius=2,color='red').add_to(m)

folium_deepnote_show(m)
```

```python
#Przykładowe użycie danych z osm

#Przystanki tramwajowe
place = 'Warsaw, PL'
tags = {'railway': 'tram_stop'}

tram_stop = ox.geometries_from_place(place, tags=tags)
tram_stop = tram_stop[tram_stop.geom_type == 'Point'][:]

m = folium.Map([52.2, 20.9], zoom_start=10)
locs = zip(tram_stop.geometry.y, tram_stop.geometry.x)
for location in locs:
    folium.CircleMarker(location=location,radius=2,color='blue').add_to(m)

folium_deepnote_show(m)
```
